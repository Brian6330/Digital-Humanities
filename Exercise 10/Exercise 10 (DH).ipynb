##### Q1: Consider the “small-network.json” file. Draw the network.
General imports and solving the question:

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import os
import json
from networkx.readwrite import json_graph
import networkx as nx
import matplotlib.pyplot as plt

with open('Data/small-network.json', 'r') as f:
    d = json.load(f)

G = json_graph.node_link_graph(d, directed=True)
print(f"Graph with {len(G.nodes())} nodes and {len(G.edges())} edges.")

nx.draw(G, with_labels=True)
plt.show()
```

#### (2) What kind of network is it?

It is an unweighted, directed network

#### (3) What is the in-degree of node d?

2, one from e and one from b.

#### (4) What is the out-degree of node c?

2, one to e and one to b.

#### (5) What is the density of this network?

```python
nx.density(G)
```

#### (6) Ignore the orientation of the nodes. What is the density of this network?

```python
G_undirected = nx.to_undirected(G)
nx.draw(G_undirected, with_labels=True)
nx.density(G_undirected)
```

#### (7) Ignore the orientation of the nodes. What is the degree of b?

```python
nx.degree(G_undirected, "b")
```

#### (8) Ignore the orientation of the nodes. Compute the mean degree of this network.

```python
print(nx.info(G_undirected))
```
